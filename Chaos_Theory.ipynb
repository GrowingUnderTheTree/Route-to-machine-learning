```python
import numpy as np
import matplotlib.pyplot as plt
```

image_2022-03-03_135559.png

```python
def lorenz(x,y,z,s=8,r=19,b=3.247):
  dotx = s*(y-x)
  doty = r*x - y - x*z
  dotz = x*y - b*z
  return dotx,doty,dotz
```

```python
dt = 0.01
numstep = 10000
```

```python
xs = np.empty(numstep + 1)
ys = np.empty(numstep + 1)
zs = np.empty(numstep + 1)
xs[0],ys[0],zs[0] = (0.,1.,1.05)
```

```python
for i in range(numstep):
  dotx,doty,dotz = lorenz(xs[i],ys[i],zs[i])
  xs[i + 1] = xs[i] + (dotx * dt)
  ys[i + 1] = ys[i] + (doty * dt)
  zs[i + 1] = zs[i] + (dotz * dt)
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
plt.show()
```
